```
from ultralytics import YOLO


# Load a model
#model = YOLO('yolov8n.yaml')  # build a new model from scratch
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

# Use the model
results = model.train(data='/content/Carla-Object-Detection-Dataset/Dataset.yaml',  name = 'CarlaCV', epochs=20, patience=20, batch = 16,
      cache = True, imgsz=640, iou = 0.5,
      augment=True, degrees=10.0, fliplr=0.0, lr0=0.001, optimizer='AdamW')  # train the model
```

```
results = model.val()  # evaluate model performance on the validation set
```
